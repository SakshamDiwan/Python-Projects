```python
#Importing the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
```

```python
#Building the list of words  

g_trends=TrendReq()
kw_list=['Artificial Intelligence','Machine Learning','Data Science']

#Receiving the Google search data of the list of words according to interest by region

g_trends.build_payload(kw_list,timeframe='today 5-y',gprop='')
data_1=g_trends.interest_by_region()

#Sorting the data in descending order according to the first element of kw_list

data_2=data_1.sort_values(by=kw_list,ascending=False)
analyse_data=data_2.head(12)

print(analyse_data)
```

```python
#Adding index to the data to make visualisation more convenient

analyse_data=analyse_data.reset_index()
print(analyse_data)
```

```python
#List creation of the different columns

list_geo=list(analyse_data['geoName'])
list_ai=list(analyse_data['Artificial Intelligence'])
list_ml=list(analyse_data['Machine Learning'])
list_ds=list(analyse_data['Data Science'])
```

```python
#Plotting the multiple bar graph chart 

fig, ax = plt.subplots(figsize=(30,20))
x=np.arange(len(list_geo))
width=0.20

ax.bar(x,list_ai,width,color="black",label='AI')
ax.bar(x+width,list_ml,width,color="green",label='ML')
ax.bar(x+width*2,list_ds,width,color="blue",label='DS')

plt.xlabel("Countries",fontweight='bold',fontsize=15)
plt.xticks(x+width,list_geo)
plt.ylabel("Number of Searches",fontweight='bold',fontsize=15)

plt.legend(prop={'size':17})
plt.show()
```

```python
g_trends1=TrendReq()
kw_list=['Artificial Intelligence','Machine Learning','Data Science']

#Receiving the Google search data of the list of words according to interest over time

g_trends1.build_payload(kw_list,timeframe='today 5-y',gprop='')
data_3=g_trends1.interest_over_time()

print(data_3)
```

```python
#Plotting the line graph 

data_3=data_3.reset_index()
fig, ax = plt.subplots(figsize=(20,15))

ax.plot(data_3['date'], data_3['Artificial Intelligence'], color="black", label="AI")
ax.plot(data_3['date'], data_3['Machine Learning'], color="green", label="ML")
ax.plot(data_3['date'], data_3['Data Science'], color="blue", label="DS")

plt.xlabel("Year",fontweight='bold',fontsize=15)
plt.ylabel("Total Count",fontweight='bold',fontsize=15)

plt.legend(prop={'size':20})
plt.show()
```
